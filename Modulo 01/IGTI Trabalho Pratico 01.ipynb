```python
#importar bibliotecas
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
#Criar o diretório para armazenar o arquivo
os.makedirs('./enem2019', exist_ok=True)
```

```python
#URL de Conexão
url = "http://download.inep.gov.br/microdados/microdados_enem_2019.zip"
```

```python
# Download
filebytes = BytesIO (
requests.get(url).content
)
# Descompactação do aquivo.
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enem2019")
```

```python
# Leitura dos dados do CSV utilizando read_csv do pandas com iterator e chunksize para quebrar o arquivo 
# e não ocupar tanto espaço em memória até utulização do filtro com os dados necessários

iter_enem2019 = pd.read_csv (
"./enem2019/DADOS/MICRODADOS_ENEM_2019.csv",
sep = ";", decimal =",", encoding = "latin1", iterator=True, chunksize=1000
)

dfMG = pd.concat([chunk[chunk['SG_UF_RESIDENCIA'] == "MG"] for chunk in iter_enem2019])
```
